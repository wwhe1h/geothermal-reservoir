```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Создаём DataFrame из твоих данных вручную
data = {
    'x': [0,0,0,1,1,1,2,2,2],
    'y': [0,1,2,0,1,2,0,1,2],
    'T (°C)': [150,148,145,147,144,140,143,139,135]
}

df = pd.DataFrame(data)

# Преобразуем DataFrame в матрицу температур (x - строки, y - столбцы)
temp_matrix = df.pivot(index='x', columns='y', values='T (°C)')

# Визуализируем тепловую карту
plt.imshow(temp_matrix, origin='lower', cmap='hot', interpolation='nearest')
plt.colorbar(label='Temperature (°C)')
plt.xlabel('y')
plt.ylabel('x')
plt.title('Temperature Heatmap')
plt.show()
```

```python
import matplotlib.animation as animation

# Зададим временные шаги
time_steps = 5

# Для примера — сгенерируем изменение температуры во времени (температура уменьшается со временем)
T_list = []
for t in range(time_steps):
    # Модель: температура падает линейно со временем (для примера)
    df['T_t'] = df['T (°C)'] - t*2  # падаем на 2 градуса на шаг
    temp_matrix_t = df.pivot(index='x', columns='y', values='T_t')
    T_list.append(temp_matrix_t.values)

# Превратим список в 3D массив
T_cube = np.array(T_list)  # форма (time_steps, x, y)

# Анимация
fig, ax = plt.subplots()
cax = ax.imshow(T_cube[0], origin='lower', cmap='hot', interpolation='nearest')
fig.colorbar(cax, label='Temperature (°C)')
ax.set_xlabel('y')
ax.set_ylabel('x')

def update(frame):
    cax.set_data(T_cube[frame])
    ax.set_title(f'Time step {frame}')
    return cax,

ani = animation.FuncAnimation(fig, update, frames=time_steps, interval=1000, blit=True)
plt.show()
```

```python
import numpy as np

# Параметры сетки
nx, ny = 3, 3  # размер сетки из примера
dx, dy = 1.0, 1.0  # шаги по координатам
dt = 0.1  # шаг по времени

# Физические параметры (пример)
alpha = 0.01  # коэффициент теплопроводности

# Начальное условие — температурный срез
T = temp_matrix.values.astype(float)  # 3x3

# Количество временных шагов
nsteps = 50

# Создаём массив для результатов
T_results = np.zeros((nsteps, nx, ny))
T_results[0] = T

for k in range(1, nsteps):
    Tn = T_results[k-1].copy()
    for i in range(1, nx-1):
        for j in range(1, ny-1):
            # Явная схема конечных разностей
            T_results[k,i,j] = (Tn[i,j] + alpha*dt*(
                (Tn[i+1,j] - 2*Tn[i,j] + Tn[i-1,j]) / dx**2 +
                (Tn[i,j+1] - 2*Tn[i,j] + Tn[i,j-1]) / dy**2
            ))
    # Граничные условия (фиксируем граничные температуры, например)
    T_results[k,0,:] = Tn[0,:]
    T_results[k,-1,:] = Tn[-1,:]
    T_results[k,:,0] = Tn[:,0]
    T_results[k,:,-1] = Tn[:,-1]
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Параметры сетки
nx, ny = 3, 3  # размер сетки
dx, dy = 1.0, 1.0  # шаги по координатам
dt = 0.1  # шаг по времени
alpha = 0.01  # коэффициент теплопроводности

# Начальное условие — температурный срез
T0 = temp_matrix.values.astype(float)  # из предыдущего шага (3x3)

nsteps = 50  # кол-во временных шагов
T_results = np.zeros((nsteps, nx, ny))
T_results[0] = T0

for k in range(1, nsteps):
    Tn = T_results[k-1].copy()
    for i in range(1, nx-1):
        for j in range(1, ny-1):
            T_results[k,i,j] = (Tn[i,j] + alpha*dt*(
                (Tn[i+1,j] - 2*Tn[i,j] + Tn[i-1,j]) / dx**2 +
                (Tn[i,j+1] - 2*Tn[i,j] + Tn[i,j-1]) / dy**2
            ))
    # Граничные условия (фиксируем граничные температуры)
    T_results[k,0,:] = Tn[0,:]
    T_results[k,-1,:] = Tn[-1,:]
    T_results[k,:,0] = Tn[:,0]
    T_results[k,:,-1] = Tn[:,-1]

# Визуализируем динамику температуры в центре сетки
center_temp = T_results[:, 1, 1]

plt.plot(np.arange(nsteps)*dt, center_temp)
plt.xlabel('Time')
plt.ylabel('Temperature at center (°C)')
plt.title('Temperature cooling at center point')
plt.grid()
plt.show()
```

```python
# Задаём исторические данные температуры в центре (искусственные)
time_history = np.arange(nsteps)*dt
true_temps = center_temp - 0.5  # чуть ниже модели, чтобы была разница
```

```python
from scipy.optimize import minimize

def solve_heat(alpha):
    T_results = np.zeros((nsteps, nx, ny))
    T_results[0] = T0
    for k in range(1, nsteps):
        Tn = T_results[k-1].copy()
        for i in range(1, nx-1):
            for j in range(1, ny-1):
                T_results[k,i,j] = (Tn[i,j] + alpha*dt*(
                    (Tn[i+1,j] - 2*Tn[i,j] + Tn[i-1,j]) / dx**2 +
                    (Tn[i,j+1] - 2*Tn[i,j] + Tn[i,j-1]) / dy**2
                ))
        # Граничные условия
        T_results[k,0,:] = Tn[0,:]
        T_results[k,-1,:] = Tn[-1,:]
        T_results[k,:,0] = Tn[:,0]
        T_results[k,:,-1] = Tn[:,-1]
    return T_results[:,1,1]  # Температура в центре

def error(alpha):
    pred = solve_heat(alpha[0])
    return np.sum((pred - true_temps)**2)

res = minimize(error, x0=[0.01], bounds=[(0.001, 0.1)])
print(f'Оптимальный alpha: {res.x[0]}')

# Строим график с оптимальным alpha
opt_alpha = res.x[0]
opt_temps = solve_heat(opt_alpha)
plt.plot(time_history, true_temps, label='Исторические данные')
plt.plot(time_history, opt_temps, label='Модель с оптимальным alpha')
plt.xlabel('Время')
plt.ylabel('Температура в центре (°C)')
plt.legend()
plt.title('Калибровка модели по истории')
plt.grid()
plt.show()
```

```python
import sys
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.backends.backend_qt5agg import FigureCanvasQTAgg as FigureCanvas
from PyQt5.QtWidgets import (
    QApplication, QWidget, QVBoxLayout, QPushButton, QSlider, QLabel, QHBoxLayout
)
from PyQt5.QtCore import Qt, QTimer

class HeatModelGUI(QWidget):
    def __init__(self):
        super().__init__()
        self.setWindowTitle("Прогноз дрейфа температуры геотермального резервуара")

        self.nx, self.ny = 30, 30  # размер сетки
        self.dx = 1.0
        self.dt = 0.1
        self.alpha = 0.01  # коэффициент теплопроводности
        self.nsteps = 200  # количество шагов по времени

        # Начальная температура
        self.T = np.full((self.nx, self.ny), 150.0)
        self.T[:, :] = 150.0
        self.T[0, :] = 150.0  # Граничные условия (пример)

        # Настройка GUI
        self.layout = QVBoxLayout()
        self.setLayout(self.layout)

        self.fig, self.ax = plt.subplots()
        self.canvas = FigureCanvas(self.fig)
        self.layout.addWidget(self.canvas)

        # Кнопки и слайдер
        btn_layout = QHBoxLayout()
        self.start_btn = QPushButton("Запустить модель")
        self.start_btn.clicked.connect(self.run_model)
        btn_layout.addWidget(self.start_btn)

        self.slider_label = QLabel("Шаг времени:")
        btn_layout.addWidget(self.slider_label)

        self.slider = QSlider(Qt.Horizontal)
        self.slider.setMinimum(0)
        self.slider.setMaximum(self.nsteps - 1)
        self.slider.valueChanged.connect(self.slider_changed)
        btn_layout.addWidget(self.slider)

        self.layout.addLayout(btn_layout)

        # Таймер для анимации
        self.timer = QTimer()
        self.timer.timeout.connect(self.update_frame)
        self.current_step = 0

        self.T_results = None
        self.im = None
        self.colorbar = None

    def solve_heat(self):
    T = self.T.copy()
    results = [T.copy()]

    for step in range(self.nsteps):
        T_new = T.copy()
        for i in range(1, self.nx - 1):
            for j in range(1, self.ny - 1):
                d2Tdx2 = (T[i+1, j] - 2*T[i, j] + T[i-1, j]) / self.dx**2
                d2Tdy2 = (T[i, j+1] - 2*T[i, j] + T[i, j-1]) / self.dx**2
                T_new[i, j] = T[i, j] + self.alpha * self.dt * (d2Tdx2 + d2Tdy2)
        T = T_new
        print(f"Step {step}: min T = {T.min()}, max T = {T.max()}")
        results.append(T.copy())
    return results


    def run_model(self):
        self.T_results = self.solve_heat()
        self.current_step = 0

        self.ax.clear()
        self.im = self.ax.imshow(self.T_results[0], cmap='hot', origin='lower')  # убрал vmin, vmax

        if self.colorbar:
            self.colorbar.remove()
        self.colorbar = self.fig.colorbar(self.im, ax=self.ax)
        self.ax.set_title("Время: 0.00, Шаг: 0")
        self.canvas.draw()

        self.slider.setValue(0)
        self.timer.start(50)  # ускоренный таймер

    def update_frame(self):
        if self.current_step >= self.nsteps:
            self.timer.stop()
            return
        frame = self.T_results[self.current_step]
        self.im.set_data(frame)
        self.ax.set_title(f"Время: {self.current_step * self.dt:.2f}, Шаг: {self.current_step}")
        self.canvas.draw()
        self.slider.blockSignals(True)
        self.slider.setValue(self.current_step)
        self.slider.blockSignals(False)
        self.current_step += 1

    def slider_changed(self, value):
        if self.T_results is None:
            return
        self.current_step = value
        frame = self.T_results[self.current_step]
        self.im.set_data(frame)
        self.ax.set_title(f"Время: {self.current_step * self.dt:.2f}, Шаг: {self.current_step}")
        self.canvas.draw()
        self.timer.stop()

def main():
    app = QApplication(sys.argv)
    window = HeatModelGUI()
    window.show()
    sys.exit(app.exec_())

if __name__ == "__main__":
    main()
```
